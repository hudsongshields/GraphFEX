```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import random
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

from FEX.models import fex
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
from generate_data import make_data, make_adjacency
timesteps=4096
adj_matrix = make_adjacency(8, 0.2, device='cpu')
data, derivatives = make_data(num_samples=timesteps, adjacency=adj_matrix)
```

```python
dimx_fex = fex.CoupledFEX('depth_2_tree_config', 'depth_3_leaves_4_config', 0, controller_epochs = 400, num_fex_epochs = 80, finetune_epochs=4000, poolsize=5, device=device, expression_threshold=0.01)
dimx_fex.fit(data, derivatives, adj_matrix, num_workers=5)
```

```python
dimy_fex = fex.CoupledFEX('depth_2_tree_config', 'depth_3_leaves_4_config', 1, controller_epochs = 500, num_fex_epochs = 80, finetune_epochs=2000, poolsize=5, device=device)
dimy_fex.fit(data, derivatives, adj_matrix, num_workers=5)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
plt.plot(data[:, :, 0].numpy(), data[:, :, 1].numpy(), c='b', alpha=0.2)
fig.axes[0].set_xlabel(r'$x_{i,1}$')
fig.axes[0].set_ylabel(r'$x_{i,2}$')
plt.show()
```
